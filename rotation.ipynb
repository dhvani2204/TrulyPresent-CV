```python
from PIL import Image
import os
import random

# Specify the root folder containing all subfolders with images
root_folder = '/content/drive/MyDrive/class_pics'

# Get list of all folders in the root folder
folders = [folder for folder in os.listdir(root_folder) if os.path.isdir(os.path.join(root_folder, folder))]

# Iterate through each folder
for folder_name in folders:
    folder_path = os.path.join(root_folder, folder_name)

    # Get list of image files in the current folder
    image_files = [f for f in os.listdir(folder_path) if f.endswith('.JPG')]

    # Iterate through each image file in the folder
    for image_file in image_files:
        image_path = os.path.join(folder_path, image_file)

        try:
            # Open the image file
            image = Image.open(image_path)

            # Check if the image size indicates a 90-degree rotation (assuming standard portrait photos)
            width, height = image.size
            if width > height:
                # Rotate counter-clockwise (90 degrees) to correct landscape orientation
                corrected_image = image.transpose(Image.ROTATE_90)
            else:
                # No correction needed for portrait orientation
                corrected_image = image

            # Apply additional random rotation (10 or 15 degrees)
            random_angle = random.randint(-15, 15) # Choose either -15 or 15 degrees randomly
            rotated_image = corrected_image.rotate(random_angle, resample=Image.BICUBIC, expand=True)

            # Save the rotated image back to the same path
            rotated_image.save(image_path)

            print(f"Corrected and rotated image {image_file} by {random_angle} degrees.")
        except Exception as e:
            print(f"Error processing {image_file}: {e}")
```
